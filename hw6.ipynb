```python
def getData():
  inFile = openFile()
  stateList = []
  popList = []
  line = inFile.readline()

  while line != "":

    #strip line of everything before and after words
    line = line.strip()
    state,pop = line.split(",")

    stateList = stateList + [state]
    popList = popList + [int(pop)]

    line = inFile.readline()
  inFile.close()
  return stateList, popList

def openFile():
  goodFile= False
  while goodFile == False:
    fname = input("Please enter the name of the data file: ")
    try:
      gradeFile = open(fname, 'r')
      goodFile = True
    except IOError:
      print("Invalid filename, please try again ... ")
  return gradeFile

#The worst case scenario for getPopLinear is if the desired state is the last on the list
def getPopLinear(stateList, popList, state):
  comps = 0
  for i in range(len(stateList)):
    comps += 1
    if stateList[i] == state:
      return comps, popList[i]
     
  return comps, -1

def selection_sort(stateList):
    n = len(stateList)
    for i in range(0, n - 1):
        min_idx = i
        for j in range(i + 1, n):
            if stateList[j] < stateList[min_idx]:
                min_idx = j
        stateList[i], stateList[min_idx] = stateList[min_idx], stateList[i]

        return stateList

#The worst case for getPopBinary is if the desired state is toward the beginning or toward the end
#of the list when it gets split down the middle
def getPopBinary(stateList,popList, state):
    stateList = selection_sort(stateList)
    start = 0
    end = len(stateList) - 1
    comps = 0

    while start <= end:
      comps += 1
      middle = (start + end) // 2
      if stateList[middle] == state:
          return comps, popList[middle]
      elif stateList[middle] < state:
          start = middle + 1
      else:
          end = middle -1
    return comps, -1

#In comparison the least possible comps for getPopLinear would be 1 while the most would be 50
#for getPoplinear it seems the least possible comps would be 4 and the most would be 6
#Binary search seems to be the better search algorithim in this case
#altough linear search performs better when it is one of the first 3 on the list

def main():
  
  stateList, popList = getData()
  state = input("Enter state abbreviation: ")
  stateList = sorted(stateList)
  linearComps, linearPop = getPopLinear(stateList, popList, state)
  binaryComps, binaryPop = getPopBinary(stateList,popList, state)
  print("Binary Search: comps = " , binaryComps)
  print("Linear Search: comps = " , linearComps)

  if linearPop or binaryPop == -1 :
    print("Invalid state abbreviation")

  elif (linearPop == binaryPop):
    pop = linearPop
    print("The population of ", state, " is " , pop)
    
  

```

```python
main()
```

```python
def getData():
  inFile = openFile()
  stateList = []
  popList = []
  line = inFile.readline()

  while line != "":

    #strip line of everything before and after words
    line = line.strip()
    state,pop = line.split(",")

    stateList = stateList + [state]
    popList = popList + [int(pop)]

    line = inFile.readline()
  inFile.close()
  return stateList, popList

def openFile():
  goodFile= False
  while goodFile == False:
    fname = input("Please enter the name of the data file: ")
    try:
      gradeFile = open(fname, 'r')
      goodFile = True
    except IOError:
      print("Invalid filename, please try again ... ")
  return gradeFile

def dualSort(popList,stateList):
    for i in range(1, len(popList)):
        save = popList[i]
        state = stateList[i]
        j = i
        while j > 0 and popList[j - 1] > save:
            # comparison
            popList[j] = popList[j - 1]
            stateList[j] = stateList[j-1]
            j = j - 1
	    # swap
        popList[j] = save
        stateList[j] = state
    return popList,stateList
def main():
    stateList, popList = getData()
    popList, stateList = dualSort(popList, stateList)
    for i in range(len(popList)):
        print(stateList[i], popList[i])
```
