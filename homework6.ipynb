```python


def poly(plaintext, key):
    ciphertext =""
    for i in range(len(plaintext)):
        # get ith letter from plaintext
        letter = plaintext[i]
        # find the position in alphabet of the letter
        pos = alphabet.index(letter)
        # find position in key of letter to shift by
        keyPos = i % len(key)
        # find letter in key to shift
        keyLetter = key[keyPos]
        # find shift value
        shift = alphabet.index(keyLetter)
        # add the shift to the position
        newPos = (pos + shift) % len(alphabet)
        # find the letter in alphabet at the new position
        newLetter = alphabet[newPos]
        # add new letter to ciphertext
        ciphertext = ciphertext + newLetter
    return ciphertext

        
        

	

```

```python
alphabet=["a","b","c","d","e","f","g","h","i","j","k","l","m","n","o","p","q","r","s","t","u","v","w","x","y","z"]

def caesarPlus(plaintext, key):
    ciphertext =""
    for i in range(len(plaintext)):
        # get ith letter from plaintext
        letter = plaintext[i]
        # find the position in alphabet of the letter
        pos = alphabet.index(letter)
        # add the key to the position
        newPos = (pos + (key+i)) % len(alphabet)
        # find the letter in alphabet at the new position
        newLetter = alphabet[newPos]
        # add new letter to ciphertext
        ciphertext = ciphertext + newLetter
    return ciphertext
def uncaesarPlus(plaintext,key):
    ciphertext =""
    for i in range(len(plaintext)):
        # get ith letter from plaintext
        letter = plaintext[i]
        # find the position in alphabet of the letter
        pos = alphabet.index(letter)
        # add the key to the position
        newPos = (pos - (key+i)) % len(alphabet)
        # find the letter in alphabet at the new position
        newLetter = alphabet[newPos]
        # add new letter to ciphertext
        ciphertext = ciphertext + newLetter
    return ciphertext
def main():
  word = caesarPlus('computerscience', 9)
  newWord = uncaesarPlus(word, 9)
  print(word)
  print(newWord)
```
